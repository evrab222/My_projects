```python
import os
import pandas as pd
import re
import warnings
import nltk
import time
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import (
    GridSearchCV,
    RandomizedSearchCV,
    train_test_split, 
    TimeSeriesSplit
)

from sklearn.feature_extraction.text import TfidfVectorizer 
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import wordnet, stopwords as nltk_stopwords

from sklearn.metrics import f1_score

from tqdm import tqdm

plt.rcParams["figure.figsize"] = (15, 10)

warnings.filterwarnings("ignore")

RANDOM_STATE = 37
TEST_SIZE = .2
```

```python
pth1 = '/datasets/toxic_comments.csv'

if os.path.exists(pth1):
    df = pd.read_csv(pth1, index_col=[0])
else:
    print('Something is wrong')
```

Загрузил данные,имеем 2 столбца *text* и *toxic* с колличеством строк равным 159292, текст комментариев на английском, есть лишние знаки, наблюдаю дисбаланс классов 10/90.

### Работа с текстом

```python
df['text'] = df['text'].str.lower()

def clear_text(text):
    clear_text = re.sub(r'[^a-z]', ' ', text)
    clear_text = clear_text.split()
    clear_text = " ".join(clear_text)
    return clear_text

df['text'] = df['text'].apply(clear_text)
```

Привел текст к нижнему регистру, очистил текст от лишних символов.

```python
nltk.download('averaged_perceptron_tagger')

def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
lemmatizer = WordNetLemmatizer()

def lemm_text(text):
    word_list = nltk.word_tokenize(text)
    lemm_text = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in word_list]
    return ' '.join(lemm_text)
```

```python
%%time
df['text'] = df['text'].apply(lemm_text)
```

Лемматизировал текст с помощью библиотеки WordNet с POS-тегом.

```python
df.tail()
```

**Вывод:** Загрузил и подготовил данные для обучения. Привел текст к нижнему регистру, очистил текст от лишних символов, так же Лемматизировал текст с помощью библиотеки WordNet с POS-тегом.

## Обучение

```python
features = df['text']
target = df['toxic']

X_train, X_test, y_train, y_test = train_test_split(features, target, 
                                                    test_size=TEST_SIZE, random_state=RANDOM_STATE,  stratify=target)
```

```python
for i in [X_train, X_test, y_train, y_test]:
    print(i.shape)
```

Разбил выборку на обучающую и тестовую,стратифицировал,вывел количество выборки после разбиения.

### Обучение LogisticRegression

```python
nltk.download('stopwords')
stopwords = set(nltk_stopwords.words('english'))
```

```python
%%time

pipeline_lr = Pipeline([("vect", TfidfVectorizer(stop_words=stopwords)), 
                     ("lr", LogisticRegression(random_state=RANDOM_STATE))])


parameters = {
    'lr__C': [0.1, 1, 10, 100],
    'lr__solver': ['liblinear', 'saga','newton-cg', 'lbfgs'],
    'lr__penalty': ['l1', 'l2'],
    'lr__class_weight': ['balanced']
}

model_lr_rs = RandomizedSearchCV(
    pipeline_lr,
    parameters,
    n_jobs=-1,
    cv=3,
    scoring='f1',
    random_state=RANDOM_STATE
)

model_lr_rs.fit(X_train, y_train) 
```

```python
results_lr = model_lr_rs.cv_results_
best_index_lr = model_lr_rs.best_index_

lr_fit_time = results_lr['mean_fit_time'][best_index_lr]
lr_score_time = results_lr['mean_score_time'][best_index_lr]
lr_f1 = (results_lr['mean_test_score'][best_index_lr])

print(f"Fit time lr: {lr_fit_time}, Predict time lr: {lr_score_time}, Best f1 lr: {lr_f1}")
```

### Обучение DecisionTreeClassifier

```python
%%time

pipeline_dtc = Pipeline([("vect", TfidfVectorizer(stop_words=stopwords)), 
                     ("dtc", DecisionTreeClassifier(random_state=RANDOM_STATE))])


parameters = {
    'dtc__min_samples_leaf': range(1, 30),
    'dtc__max_depth': range(2, 10),
    'dtc__class_weight': ['balanced']
} 

model_dtc_rs = RandomizedSearchCV(
    pipeline_dtc,
    parameters,
    n_jobs=-1,
    cv=3,
    scoring='f1',
    random_state=RANDOM_STATE,
    error_score='raise'
)

model_dtc_rs.fit(X_train, y_train) 
```

```python
results_dtc = model_dtc_rs.cv_results_
best_index_dtc = model_dtc_rs.best_index_

dtc_fit_time = results_dtc['mean_fit_time'][best_index_dtc]
dtc_score_time = results_dtc['mean_score_time'][best_index_dtc]
dtc_f1 = (results_dtc['mean_test_score'][best_index_dtc])

print(f"Fit time dtc: {dtc_fit_time}, Predict time dtc: {dtc_score_time}, Best f1 dtc: {dtc_f1}")
```

### Обучение KNeighborsClassifier

```python
%%time

pipeline_knc = Pipeline([("vect", TfidfVectorizer(stop_words=stopwords)), 
                     ("knc", KNeighborsClassifier())])

parameters = {'knc__n_neighbors' : [2, 5, 10]}

model_knc_rs = RandomizedSearchCV(
    pipeline_knc,
    parameters,
    n_jobs=-1,
    cv=3,
    scoring='f1',
    random_state=RANDOM_STATE
)

model_knc_rs.fit(X_train, y_train) 
```

```python
results_knc = model_knc_rs.cv_results_
best_index_knc = model_knc_rs.best_index_

knc_fit_time = results_knc['mean_fit_time'][best_index_knc]
knc_score_time = results_knc['mean_score_time'][best_index_knc]
knc_f1 = (results_knc['mean_test_score'][best_index_knc])

print(f"Fit time knc: {knc_fit_time}, Predict time knc: {knc_score_time}, Best f1 knc: {knc_f1}")
```

### Обучение RandomForestClassifier

```python
%%time

pipeline_rfc = Pipeline([("vect", TfidfVectorizer(stop_words=stopwords)), 
                     ("rfc", RandomForestClassifier(random_state=RANDOM_STATE))])

parameters = {
    'rfc__n_estimators': range(2, 60, 5),
    'rfc__max_depth': range(2, 10),
    'rfc__min_samples_leaf': range(1,5),
    'rfc__min_samples_split': range(2,6),
    'rfc__class_weight': ['balanced']
}

model_rfc_rs = RandomizedSearchCV(
    pipeline_rfc,
    parameters,
    n_jobs=-1,
    cv=3,
    scoring='f1',
    random_state=RANDOM_STATE
)

model_rfc_rs.fit(X_train, y_train) 
```

```python
results_rfc = model_rfc_rs.cv_results_
best_index_rfc = model_rfc_rs.best_index_

rfc_fit_time = results_rfc['mean_fit_time'][best_index_rfc]
rfc_score_time = results_rfc['mean_score_time'][best_index_rfc]
rfc_f1 = (results_rfc['mean_test_score'][best_index_rfc])

print(f"Fit time rfc: {rfc_fit_time}, Predict time rfc: {rfc_score_time}, Best f1 rfc: {rfc_f1}")
```

### Обучение CatBoostClassifier

```python
X_train_cbc = X_train[:1000]
X_test_cbc = X_test[:200]
y_train_cbc = y_train[:1000]
y_test_cbc = y_test[:200]
```

Ограничил колличество строк для обучения CatBoostClassifier,при всем объеме df ядро слетает.

```python
%%time

pipeline_cbc = Pipeline([("vect", TfidfVectorizer(stop_words=stopwords)), 
                     ("cbc", CatBoostClassifier(random_state=RANDOM_STATE))])

# model_cbc = CatBoostClassifier(random_state=RANDOM_STATE, verbose=False)
parameters = {'cbc__depth' : [2, 5, 10]}

model_cbc_gs = GridSearchCV(pipeline_cbc, 
                            parameters, 
                            cv = 3,
                            n_jobs=-1, 
                            scoring='f1'
)

model_cbc_gs.fit(X_train_cbc, y_train_cbc)
```

```python
results_cbc = model_cbc_gs.cv_results_
best_index_cbc = model_cbc_gs.best_index_

cbc_fit_time = results_cbc['mean_fit_time'][best_index_cbc]
cbc_score_time = results_cbc['mean_score_time'][best_index_cbc]
cbc_f1 = (results_cbc['mean_test_score'][best_index_cbc])

print(f"Fit time cbc: {cbc_fit_time}, Predict time cbc: {cbc_score_time}, Best f1 cbc: {cbc_f1}")
```

```python
result_train = pd.DataFrame(
    [[lr_fit_time, lr_score_time, lr_f1],
    [dtc_fit_time, dtc_score_time, dtc_f1],
    [knc_fit_time, knc_score_time, knc_f1],
    [rfc_fit_time, rfc_score_time, rfc_f1],
    [cbc_fit_time, cbc_score_time, cbc_f1]],
    index=['LogisticRegression', 'DecisionTreeClassifier',
           'KNeighborsClassifier', 'RandomForestClassifier', 'CatBoostClassifier'],
    columns=['fit_time', 'predict_time', 'f1']
)
result_train
```
