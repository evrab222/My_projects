В данной таблице очень много пропусков,разберемся с каждым столбцом и поработаем с пропущенными значениями.


Удалим пропущенные значения в двух столбцах,заменить их нечем и на общий анализ не повлияет,так как сумма пропусков составляет лишь 0,5% от всех строк.

```python
data.dropna(subset=['locality_name'], inplace=True)

data.dropna(subset=['floors_total'], inplace=True)
#удаляем строки с пропусками
```

Поработаем со столбцом 'ceiling_height'.

```python
data['ceiling_height'].sort_values(ascending=False).head(20)
```

Прежде чем заменять пропуски на мединные значения,разберемся с аномальными значениями.
Возможно значения от 21 до 32 были написаны по ошибке и логично предположить, что на самом деле это вещественные значения 2.1,3.2 и тд.

```python
data = data[data['ceiling_height'] != 100]
#отфильтруем и уберем значение 100 
data['ceiling_height'] = data['ceiling_height'].apply(lambda x: x / 10 if x > 21 else x)
#заменим значения от 21 до 32 на значения деленные на 10
```

Заменим пропущенные значения в стобце 'ceiling_height' на среднее.Прежде вычеслим среднее,отбросив значения до 2.1 и после 6.0(в жилих помещениях ниже 2.1 нельзя жить по нормам СНиП, и выше 6 метров тоже аномальные значения),получаем значение 2.7.

```python
data.query('ceiling_height >= 2.1 & ceiling_height <= 6.0')['ceiling_height'].median()
#вычесляем среднюю,отфильтровав значения 
data.loc[data['ceiling_height'] >= 6.0, 'ceiling_height'] = 2.7
#заменяем значения от 6.0 на среднее
data.loc[data['ceiling_height'] <= 2.1, 'ceiling_height'] = 2.7
#заменяем значения до 2.1 на среднее
data['ceiling_height'] = data['ceiling_height'].fillna(2.7)
#заменяем пропуски на среднее значение

data.reset_index(drop= True , inplace= True)
#сбрасываем и удаляем индексы
```

### Посчитайте и добавьте в таблицу новые столбцы

#### Добавление столбца цена за 1 кв.м.

```python
data['price_one_metr'] = data['last_price'] / data['total_area']
data['price_one_metr'] = data['price_one_metr'].astype(int)
```

#### Добавление столбцов день недели,месяц,год публикации обьявлений.

```python
data['day_of_week'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['years'] = data['first_day_exposition'].dt.year
```

#### Добавление столбца с типом этажа.

```python
def floor_function(floor):
    if floor['floor'] == 1:
        return 'первый'
    elif floor['floor'] == floor['floors_total']:
        return 'последний'
    else:
        return 'другой'

data['floor_type'] = data.apply(floor_function, axis=1)
#использовал функцию для градации типов этажа
```

#### Добавление столбца с растоянием до центра города в километрах.

```python
data['city_centers_km'] = data['city_centers_nearest'].apply(lambda x: x if x == (-1) else x / 1000).astype(int)
#отбросил маркеры и добавил столбец в км,сделал целочисленным столбец
```
